<font color='skyblue'> Tim_Sort </font> is the hy<br>id of <font color='green'> Merge_Sort </font> and <font color='yellow'> Insertion_Sort </font>

```python
def Insertion_sort(arr,start,end):
    for i in range(start+1,end+1):
        j=i-1
        key=arr[i]

        while j>=start and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
```

```python
def Merge_sort(arr,left,mid,right):
    le=arr[left:mid+1]
    ri=arr[mid+1:right+1]

    j=i=0
    k=left
    while i<len(le) and j<len(ri):
        if le[i]<=ri[j]:
            arr[k]=le[i]
            i+=1
        else:
            arr[k]=ri[j]
            j+=1
        k+=1
    
    while i<len(le):
        arr[k]=le[i]
        i+=1
        k+=1
    while j<len(ri):
        arr[k]=ri[j]
        j+=1
        k+=1


```

```python
def Tim_sort(arr):
    n=len(arr)
    Run=32
    size=Run

    for i in range(0,n,Run):
        Insertion_sort(arr,i,min(i+Run-1,n-1))
    
    while size<n:
        for left in range(0,n,2*size):
            mid=min(n-1,left+size-1)
            right=min(n-1,left+2*size-1)

            if mid<right:
                Merge_sort(arr,left,mid,right)
        size*=2
    return arr
```

```python
arr=[3,5,3,2,5,7]
Tim_sort(arr)
```
